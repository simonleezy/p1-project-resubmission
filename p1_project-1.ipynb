```python
# convert relevant columns to numeric
numeric_columns = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
#replacing unknown in broad phase of flight
df['Broad.phase.of.flight'].replace('unknown', np.nan, inplace=True)
```

```python
#drop NaN values
df.dropna(subset=['Broad.phase.of.flight'], inplace=True)
```

```python
#grouping by broad phase of fight and sum of injury
df_grouped = df.groupby('Broad.phase.of.flight')[numeric_columns].sum().reset_index()
```

```python
#format for easie plotting
df_melted = pd.melt(df_grouped, id_vars=['Broad.phase.of.flight'],
                    value_vars=numeric_columns,
                    var_name='Injury Type', value_name='count')
```

```python
#GROUPED BAR PLOT
plt.figure(figsize=(14,8))
sns.barplot(x='Broad.phase.of.flight', y='count', hue='Injury Type', data=df_melted)
plt.xlabel('Broad.phase.of.flight')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.legend(title='Injury Type')

plt.tight_layout()
plt.show()
```

```python
#WHERE Broad phase of flight is unknown
filtered_df = df[df['Broad.phase.of.flight'] !='unknown']
```

```python
#NUMBER OF ACCIDETS IN BROAD PHASE OF FLIGHT
phase_counts = filtered_df['Broad.phase.of.flight'].value_counts()
```

```python
#BAR CHART
plt.figure(figsize=(12, 6))
phase_counts.plot(kind='bar', color='blue')
plt.title('Number of Accidents')
plt.xlabel('Broad Phase of Flight')
plt.ylabel('Number of Accidents')
plt.grid(axis='y')
plt.show()
```
